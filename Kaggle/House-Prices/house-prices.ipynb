```python
df3 = pd.concat([df2, test],)
#df3.iloc[:1460, :]
df3
```

```python
columns = []
for x in df3.columns:
    if df3[x].dtype == 'O':
        columns.append(x)
        
for column in columns:
    dummies = pd.get_dummies(df3[column])
    dummies.drop(columns=dummies.columns[-1], inplace=True)
    df3[dummies.columns] = dummies

df3.drop(columns=columns, inplace=True)
```

```python
train_data = df3.iloc[:1460, :]
test_data = df3.iloc[1460:, :]
test_data.info()
```

```python
X_columns = []
X_remove_columns = ['SalePrice', 'Id']
for _ in train_data.columns:
    if _ in X_remove_columns:
        pass
    else:
        X_columns.append(_)
X = train_data[X_columns]
y = train_data['SalePrice']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
model = LinearRegression().fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
X_columns = []
for Q in test_data.columns:
    if Q in X_remove_columns:
        pass
    else:
        X_columns.append(Q)
X_predict = test_data[X_columns]
```

```python
price_predict = model.predict(X_predict)
```
